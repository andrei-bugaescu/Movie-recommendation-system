```python
def import_database(file_path,file_name, separator,columns_names):
  import pandas as pd
  #import data (csv)
  #header : 0 => the first line of the file =  column names
  file_ = file_path+file_name
  df = pd.read_csv(file_, compression='gzip', header=0,sep=separator, quotechar='"', error_bad_lines=False)
  #columns renaming
  #df.set_axis(columns_name, axis=1, inplace=False)
  df = df.set_axis(columns_names, axis=1, inplace=False)
  return df
```
